```python
class Calculator:
    def add(n1,n2):
        print("Addition:",n1+n2)

    def sub(n1,n2):
        print("Subtraction:",n1-n2)

    def multiply(n1,n2):
        print("Multiply:",n1*n2)

    def divison(n1,n2):
        print("Division:",n1//n2)
```

```python
n1=int(input("Enter no 1:"))
n2=int(input("Enter no 2:"))

c1=Calculator
c1.add(n1,n2)
c1.sub(n1,n2)
c1.multiply(n1,n2)
c1.divison(n1,n2)
```
