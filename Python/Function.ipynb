```python
def billing_system(amount, gst, quantity, discount):
    # Step 1: Add GST to amount
    gst_amount = amount * (gst / 100)
    amount_with_gst = amount + gst_amount

    # Step 2: Multiply by quantity
    total_amount = amount_with_gst * quantity

    # Step 3: Apply discount
    discount_amount = total_amount * (discount / 100)
    final_amount = total_amount - discount_amount

    print('GST :',gst,'%')
    print('GSTAmount :',gst_amount,'₹')
    print('Amount With GST :',amount_with_gst,'₹')
    print('TotalAmount :',total_amount,'₹')
    print('Dicount Amount :',discount_amount,'₹')
    print('Finalprice :',final_amount,'₹')

billing_system(1000,12,5,10)
```
