```python
import matplotlib.pyplot as plt

# sample data

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(y,x, color='Red', linestyle='--', marker='o', markersize=8)

# add labels and title
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Simple line plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# sample data

x = [2,2,3,20,2]
y = [2,4,6,3,10]
z = [20,40,60,30,10]

plt.plot(y,x, color='Red', linestyle='--', marker='o', markersize=8)
plt.plot(y,z, color='Red', linestyle='--', marker='o', markersize=8)


# add labels and title
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Simple line plot')
plt.show()
```

```python
cat = ['java','c++','python','go']
value = [20,30,50,10]

plt.bar(cat,value,color='Green')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Simple line plot')
plt.show()
```

```python
# adding grid

plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('simple line plot')
plt.grid(True)
plt.show()
```

```python
# scatter plot 

plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('simple line plot')
plt.show()
```

```python
import numpy as np
np.random.seed(43)
data = np.random.normal(loc=170,scale=10,size=1000) # random data creation

plt.hist(data,bins=30,color='skyblue',edgecolor='black') # histogram creation
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Simple line plot')
plt.show()
```
