```python
# pd.DataFrame

students = {
    "name":['joy','Nishant','ronak','tiya','riya'],
    "age":[23,45,67,12,22],
    "city":['mumbai','pune','jaipur','bhuj','surat']
}

df = pd.DataFrame(students)
print(df)

# change index 

indexx = [101,102,103,104,105]

df = pd.DataFrame(students,index=indexx)
print(df)

df.to_csv('student.csv',index=False)
```
