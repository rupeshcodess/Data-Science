```python
import pandas as pd

data = {
    'Store': ['North', 'South', 'North', 'East', 'South', 'East', 'North'],
    'Product': ['Laptop', 'Phone', 'Laptop', 'Tablet', 'Phone', 'Laptop', 'Tablet'],
    'Sales': [1200, 800, 1500, 600, 900, 1100, 700],
    'Quantity': [2, 3, 3, 1, 4, 2, 2]
}

df = pd.DataFrame(data)
print(df)
```

```python
# group by store and sum all numberic columns
store_summary = df.groupby('Store').sum()
print(store_summary)
```

```python
# group single column
sales_by_store = df.groupby('Store')['Sales'].sum()
print(sales_by_store)
```

```python
# multiple Aggregation (sum,mean,count)
summary = df.groupby('Store').agg({
    "Sales":['sum','mean','count'],
    "Quantity":'sum'
})

print(summary)
```

```python
# group by multiple column
by_store_product = df.groupby(['Store','Product'])['Sales'].sum()
print(by_store_product)
```

```python
data = pd.read_csv("credit.csv")
print(data.head())
```

```python
amount_trans = data.groupby('Transaction_Type')['Amount'].sum()
print(amount_trans)
```

```python
amount_mer = data.groupby('Merchant')['Amount'].sum()
print(amount_mer)
```

```python
amount_cat = data.groupby(['Merchant','Category'])['Amount'].sum()
print(amount_cat)
```

```python
summary_amount = data.groupby('Merchant').agg({
    "Amount":['sum','mean'],
    "Category":'sum',
    "Transaction_Type":'sum'
})

print(summary_amount)
```
