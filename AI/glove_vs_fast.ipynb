```python
import gensim.downloader as api

glove = api.load("glove-wiki-gigaword-300")
```

```python
# wikipedia and english fast text
fast = api.load("fasttext-wiki-news-subwords-300")
```

```python
# interface
print(glove.most_similar("movie"))
print(fast.most_similar("movie"))
```

```python
# analogy
print(glove.most_similar(positive=["movie", "film"], negative=["tv"]))
print(fast.most_similar(positive=["movie", "film"], negative=["tv"]))
```
