```python
# convert Transaction_Date to date format
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'],errors='coerce')
```

```python
# check datatype
df.dtypes
```

```python
# add year , month , and day 
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Day'] = df['Transaction_Date'].dt.day
```

```python
df.head()
```

```python
# check info
df.info()
```

```python
# find the Transactions january 2023 
jan_23 = df[(df['Year'] == 2023)& (df['Month']==1)]
print(jan_23.head())
```

```python
# find the transactions with amount > 2500 transaction type = online 
online_1000 = df[(df['Amount'] > 2500) & (df['Transaction_Type']=='Online')]
# print(online_1000)

# create csv 
online_1000.to_csv('online2500.csv')
```

```python
df['Discounted_Amount'] = df['Amount'].apply(lambda x: x * 0.95 if x > 500 else x)
print(df[['Amount', 'Discounted_Amount']].head())
```

```python
df['Amount_Category'] = pd.cut(df['Amount'], 
                               bins=[0, 100, 500, float('inf')],
                               labels=['Low', 'Medium', 'High'])
print(df[['Amount', 'Amount_Category']].head())
```

```python
missing_merchant_percentage = df['Merchant'].isnull().mean()
if missing_merchant_percentage > 0.3:
    df.drop(columns=['Merchant'], inplace=True)
    print(f"\nMerchant column dropped due to {missing_merchant_percentage:.2%} missing values.")
else:
    print(f"\nMerchant column retained ({missing_merchant_percentage:.2%} missing values).")
```

```python
# total tranaction amount per category
total_amount_per_category = df.groupby('Category')['Amount'].sum().reset_index()
print(total_amount_per_category)
```

```python
# number of declined transactions per Payment_Mode.
declined_per_payment_mode = df[df['Transaction_Status'] == 'Declined'].groupby('Payment_Mode').size().reset_index(name='Declined_Count')
print(declined_per_payment_mode)
```

```python
# top 5 most frequent merchants
if 'Merchant' in df.columns:
    top_merchants = df['Merchant'].value_counts().head(5).reset_index()
    top_merchants.columns = ['Merchant', 'Transaction_Count']
    print("\nTop 5 Most Frequent Merchants:")
    print(top_merchants)
else:
    print("\nMerchant column not available for analysis.")
```

```python
# average transaction amount per Location.
avg_amount_per_location = df.groupby('Location')['Amount'].mean().reset_index()
print(avg_amount_per_location)
```

```python
# Customers with >10 transactions in a single day
df['Date'] = df['Transaction_Date'].dt.date
transactions_per_day = df.groupby(['Customer_ID', 'Date']).size().reset_index(name='Transaction_Count')
potential_fraud_customers = transactions_per_day[transactions_per_day['Transaction_Count'] > 10]
print("Customers with >10 Transactions in a Single Day (Potential Fraud):")
print(potential_fraud_customers)
```

```python
# Transactions with same Customer_ID in different locations within 5 minutes
df_sorted = df.sort_values(['Customer_ID', 'Transaction_Date'])
df_sorted['Time_Diff'] = df_sorted.groupby('Customer_ID')['Transaction_Date'].diff().dt.total_seconds() / 60
df_sorted['Location_Change'] = df_sorted.groupby('Customer_ID')['Location'].shift() != df_sorted['Location']
same_customer_diff_location = df_sorted[(df_sorted['Time_Diff'].notnull()) & (df_sorted['Time_Diff'] <= 5) & (df_sorted['Location_Change'])]
print("Transactions with Same Customer_ID in Different Locations within 5 Minutes:")
print(same_customer_diff_location[['Customer_ID', 'Transaction_Date', 'Location', 'Time_Diff']])
```

```python
# High-risk transactions (Amount > $5000 and Online)
high_risk_transactions = df[(df['Amount'] > 5000) & (df['Transaction_Type'] == 'Online')]
print("High-Risk Transactions (Amount > $5000 and Online):")
print(high_risk_transactions)
```
