```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Load dataset
data = pd.read_csv("Mobile phone price.csv")

# Create Category column (0 = Cheap, 1 = Expensive)
data["Category"] = np.where(data["Price ($)"] <= data["Price ($)"].median(), 0, 1)

# Select features
X = data[["Price ($)", "Battery Capacity (mAh)"]].values
y = data["Category"].values

# Train KNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# Predict
oneplus_nord = np.array([[300, 4500]])
prediction = knn.predict(oneplus_nord)

categories = {0: "Cheap", 1: "Expensive"}
print("Prediction:", categories[prediction[0]])
```
