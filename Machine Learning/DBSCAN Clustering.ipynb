```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# data point 
points = np.array([
    [1,1],[1.5,1.5],[2,1],[5,5],[5.5,5.5],[6,5],[5.5,6],[6.5,6.5],[10,10],[2.5,2.5],[2,2],[3,3]
])

# apply DBSCAN

db = DBSCAN(eps=1.9,min_samples=4).fit(points)

# set cluster labels (-1 indicates noise )
labels = db.labels_

# visualize
plt.figure(figsize=(8,6))
unique_labels = set(labels)
colors = ['r','b','g','y']+['k']

for k,col in zip(unique_labels,colors[:len(unique_labels)]):
    if k == -1:
        col = 'k' # black for noise
    class_member = (labels == k)
    xy = points[class_member]
    plt.plot(xy[:, 0],xy[:, 1],'o',markerfacecolor=col,markeredgecolor='k',markersize=10,label=f'cluster {k}' if k != -1 else 'Noise')
plt.title("DBSCAN ---- Data Point")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True)
plt.show()
```
